```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

#1. Histogram of BMI Distribution
This histogram visualizes the distribution of Body Mass Index (BMI) among individuals in the Diabetes dataset, with a kernel density estimate (KDE) to show the distribution curve. Understanding BMI distribution is crucial as BMI is a significant risk factor for diabetes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


diabetes_data = pd.read_csv('Diabetes.csv')

plt.figure(figsize=(10, 6))
sns.histplot(diabetes_data['BMI'], kde=True, color='skyblue')
plt.title('Distribution of BMI in Diabetes Dataset')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()

```

#Scatter Plot Graph:
 It shows the relationship between age and plasma glucose concentration (PG Concentration), with the points colored based on the diabetes classification (Sick or Healthy). This visualization helps to understand if there's any visible trend or pattern between age, glucose concentration, and the diabetes classification within the dataset. ​​








```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='PG Concentration', hue='class', data=diabetes_data, palette='coolwarm')
plt.title('Age vs. PG Concentration by Diabetes Class')
plt.xlabel('Age')
plt.ylabel('PG Concentration')
plt.legend(title='Diabetes Class')
plt.show()
```

#Box Plot of BMI by Diabetes Class:
The box plot shows the distribution of BMI across different diabetes classifications (Sick or Healthy). It allows for quick identification of the median, quartiles, and potential outliers in BMI within each class, offering insights into the relationship between BMI and diabetes.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='BMI', data=diabetes_data, palette='pastel')
plt.title('BMI Distribution by Diabetes Class')
plt.xlabel('Diabetes Class')
plt.ylabel('BMI')
plt.show()
```

#Violin Plot of Age by Diabetes Class:
This violin plot illustrates the distribution of ages within each diabetes classification, providing a richer understanding of the age distribution with both the density estimation and the interquartile range visible. It is useful for observing age-related patterns in the context of diabetes classification.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='class', y='Age', data=diabetes_data, palette='autumn')
plt.title('Age Distribution by Diabetes Class')
plt.xlabel('Diabetes Class')
plt.ylabel('Age')
plt.show()
```
